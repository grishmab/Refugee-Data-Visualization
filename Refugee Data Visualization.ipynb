# A

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

data1 = pd.read_csv("~/share/Datasets/refugee_status_trimmed.csv", index_col="Continent")
data1
```

### The first plot is a simple line plot of total number of refugees over time.

```python
total = data1.loc["Total",:]
```

```python
lp_fig, lp_ax = plt.subplots(figsize=(10,7))
plt.plot(total)
plt.title("Total Refugees Over Time (2006-2015)")
plt.xlabel("Year")
plt.ylabel("Total Number of Refugees");
```

### The second plot is a line plot of refugees over time by continent.

```python
mlp_fig, mlp_ax = plt.subplots(figsize=(10,7))
africa = data1.loc["Africa",:]
plt.plot(africa)
asia = data1.loc["Asia",:]
plt.plot(asia)
europe = data1.loc["Europe",:]
plt.plot(europe)
namerica = data1.loc["North America",:]
plt.plot(namerica)
samerica = data1.loc["South America",:]
plt.plot(samerica)
uknown = data1.loc["Unknown",:]
plt.plot(uknown)
mlp_ax.legend(loc='upper left')
plt.title("Refugees Over Time (2006-2015) by Continent")
plt.xlabel("Year")
plt.ylabel("Total Number of Refugees");
```

### The third plot is a stacked line plot of refugees over time by continent.

```python
slp_fig, slp_ax = plt.subplots(figsize=(10,7))
africa = data1.loc["Africa",:]
asia = data1.loc["Asia",:]
europe = data1.loc["Europe",:]
namerica = data1.loc["North America",:]
samerica = data1.loc["South America",:]
uknown = data1.loc["Unknown",:]
x = np.arange(2006,2016)
lbls = ["Africa","Asia","Europe","NAmerica","SAmerica","Unknown"]
plt.stackplot(x, africa, asia, europe, namerica, samerica, uknown, labels=lbls)
slp_ax.legend(loc='upper left')
plt.title("Refugees Over Time (2006-2015) by Continent")
plt.xlabel("Year")
plt.ylabel("Total Number of Refugees");
```

### The fourth and final plot for this question is a clustered bar chart of refugees over time by continent.

```python
barWidth = 0.18
cbc_fig, cbc_ax = plt.subplots(figsize=(10,7))
 
africa = data1.loc["Africa",:]
asia = data1.loc["Asia",:]
europe = data1.loc["Europe",:]
namerica = data1.loc["North America",:]
samerica = data1.loc["South America",:]
 
# Set position of bar on X axis
r1 = np.arange(len(africa))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
 
# Make the plot
plt.bar(r1, africa, width=barWidth, edgecolor='white', label='Africa')
plt.bar(r2, asia, width=barWidth, edgecolor='white', label='Asia')
plt.bar(r3, europe, width=barWidth, edgecolor='white', label='Europe')
plt.bar(r4, namerica, width=barWidth, edgecolor='white', label='NAmerica')
plt.bar(r5, samerica, width=barWidth, edgecolor='white', label='SAmerica')

# Add xticks on the middle of the group bars
plt.title("Refugees Over Time (2006-2015) by Continent")
plt.xlabel("Year")
plt.ylabel("Total Number of Refugees");
plt.xticks([r + barWidth for r in range(len(africa))], ['2006', '2007', '2008', '2009', '2010','2011','2012','2013','2014','2015'])
 
cbc_ax.legend(loc='upper left')
plt.show()
```

```python
data2 = pd.read_csv("~/DS & V Classwork/refugee_statistics.csv", index_col = "Sex/Age/Marital Status")
```

### B) Using the refugee_stats data, which contains data for 2015 refugees, generate several visualizations. First, a clustered bar chart conveying the sex (as identified by the dataset) of the refugees, both applicants and all refugees. Do the same, with the initial first three categories of ages (under 16, 16-20, 21 and over). Last, create both a standard line plot and a stacked line plot of the age of applicants and all refugees across the more granular age groups (under 1 through 75 and over).

### The first plot for this question is a clustered bar chart conveying the sex as identified by the dataset of the refugees, both applicants and all refugees. 

```python
stats1 = data2.iloc[0:3, :]

width = 0.15 
fig, ax = plt.subplots(figsize=(12,8))

female = stats1.loc["Female",:]
male = stats1.loc["Male",:]
unknown = stats1.loc["Sex Unknown",:]

y1 = np.arange(len(female))
y2 = [x + width for x in y1]
y3 = [x + width for x in y2]

plt.bar(y1, female, width, label='Female')
plt.bar(y2, male, width, label='Male')
plt.bar(y3, unknown, width, label='Unknown')

plt.title("Data for Refugees in 2015 According to Sex")
plt.ylabel('Number of Refugees')
plt.xticks([y + width for y in range(len(female))],['Applicants','All Refugees'])
ax.legend(loc = 'upper left')
plt.show()
```

### The second plot for this question is a clustered bar chart conveying the first three categories of age as identified by the dataset of the refugees, both applicants and all refugees. 

```python
stats2 = data2.iloc[3:6, :]

width = 0.15 
fig, ax = plt.subplots(figsize=(12,8))

a = stats2.loc["Under 16 Years",:]
b  = stats2.loc["16 to 20 Years",:]
c = stats2.loc["21 Years and Over",:]

y1 = np.arange(len(a))
y2 = [x + width for x in y1]
y3 = [x + width for x in y2]

plt.bar(y1, a, width, label='Under 16 Years')
plt.bar(y2, b, width, label='16 to 20 Years')
plt.bar(y3, c, width, label='21 Years and Over')

plt.title("Data for refugees in 2015 according to age")
plt.ylabel("Number of refugees")
plt.xticks([y + width for y in range(len(a))],['Applicants','All Refugees'])
ax.legend(loc = 'upper left')
plt.show()
```

### The third and final plot for this question are a standard line plot and a stacked line plot of the age of applicants and all refugees across the more granular age groups (under 1 through 75 and over).

### a. Standard Line Plot

```python
stats3 = data2.iloc[6:22, :]
lp_fig, lp_ax = plt.subplots(figsize=(22,8))
app =stats3.loc[:,"Applicants"]
plt.plot(app)
all_ref = stats3.loc[:,"All Refugees"]
plt.plot(all_ref)
lp_ax.legend(loc = 'upper left')
plt.title("Standard line plot of refugees (under 1 through 75 and over)")
plt.xlabel("Years")
plt.ylabel("Number of refugees");
```

### b. Stacked Line Plot

```python
slp_fig, slp_ax = plt.subplots(figsize=(12,8))
x = np.arange(len(stats3))

lbls = ['Applicants','All Refugees']
slp_ax.stackplot(x, app, all_ref, labels = lbls)
slp_ax.set_title("Standard line plot of refugees (under 1 through 75 and over)")
slp_ax.set_ylabel("Number of refugees")
slp_ax.set_xlabel("Years; 0-15 is (under 1)-(75 over) respectively")
slp_ax.set_xticks(x)

slp_ax.legend(loc = 'upper left')
plt.show()
```
